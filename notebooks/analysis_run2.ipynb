```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

%matplotlib inline
```

```python
df = joblib.load('../data/clean_data.pkl')
# display un-truncated columns when calling upon dataframe
pd.set_option('display.max_columns', None)
```

```python
#Subsetting Data Based off Loan Status
df_charged_off = df.query('loan_status == "Charged Off"')
df_fully_paid = df.query('loan_status == "Fully Paid"')
```

```python
# List of numeric columns from dataframe
numeric_cols = df.select_dtypes(include = [int, float]).columns
```

```python
def kdeplot_vs_repay(col, df1, df2):
    sns.kdeplot(data= df1, x=col, color='red', alpha=0.5, common_grid = True, fill = True)
    sns.kdeplot(data=df2, x=col, color='blue', alpha=0.5, common_grid = True, fill = True)
    plt.legend(labels=['Repayed', 'Failed to Repay'])
    plt.show()
```

```python
num_cols_of_interest = ['int_rate','dti' ]
```

```python
for i in numeric_cols:
    kdeplot_vs_repay(i, df_fully_paid.query('annual_inc < 40000'), df_charged_off.query('annual_inc < 40000'))
```

```python
for i in numeric_cols:
    kdeplot_vs_repay(i, df_fully_paid.query('annual_inc > 40000 & annual_inc < 820000'), df_charged_off.query('annual_inc > 40000 & annual_inc < 820000'))
```

```python
for i in numeric_cols:
    kdeplot_vs_repay(i, df_fully_paid.query('annual_inc > 82000 & annual_inc < 650000'), df_charged_off.query('annual_inc > 82000 & annual_inc < 650000'))
```
